```python
import pandas as pd
import matplotlib.pyplot as plt

# Prepare the DataFrame
data = {
    "Model": [
        "LWM_freeze_backbone",
        "LWM_pretrained_Fine_tune",
        "LWM_Fine_tune",
        "GRU",
        "RNN",
        "LSTM",
        "Transformer"
    ],
    "Best Val NMSE (dB)": [-14.51, -16.50, -16.45, -15.79, -16.34, -14.51, -16.33],
    "Avg Time per Epoch (s)": [203.26, 255.22, 262.35, 101.57, 82.70, 109.80, 191.74]
}

df = pd.DataFrame(data)

# Scatter plot: Avg Time vs. Best Val NMSE
plt.figure(figsize=(8, 5))
plt.scatter(df["Avg Time per Epoch (s)"], df["Best Val NMSE (dB)"])

# Annotate each point with its model name
for _, row in df.iterrows():
    plt.text(row["Avg Time per Epoch (s)"], row["Best Val NMSE (dB)"], row["Model"], fontsize=8, ha='right', va='bottom')

plt.title("Trade-off Between Validation NMSE and Training Time")
plt.xlabel("Average Time per Epoch (s)")
plt.ylabel("Best Validation NMSE (dB)")
plt.gca().invert_yaxis()  # More negative (better) NMSE at the top
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data: epochs 1-10
epochs = list(range(1, 11))

nmse = {
    "LWM_freeze_backbone": [-13.4, -13.6, -13.8, -13.9, -14.0, -14.1, -14.2, -14.3, -14.4, -14.5],
    "LWM_pretrained_Fine_tune": [-14.0, -14.9, -15.3, -15.4, -15.6, -16.0, -16.2, -16.3, -16.4, -16.5],
    "LWM_Fine_tune": [-15.3, -16.0, -15.9, -15.8, -15.8, -16.0, -16.2, -16.3, -16.3, -16.4],
    "GRU": [-13.2, -13.9, -14.2, -15.3, -15.5, -15.6, -15.6, -15.6, -15.7, -15.8],
    "RNN": [-14.3, -15.7, -16.0, -16.0, -16.1, -16.0, -16.1, -16.2, -16.3, -16.3],
    "LSTM": [-13.4, -13.4, -13.8, -13.9, -14.2, -14.0, -14.2, -13.6, -14.4, -14.5],
    "Transformer": [-15.3, -15.6, -15.6, -15.6, -15.9, -15.9, -16.1, -16.3, -16.3, -16.3]
}

plt.figure(figsize=(10, 7))

for model, values in nmse.items():
    plt.plot(epochs, values, marker='o', label=model)

plt.title("Validation NMSE(dB) over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Val NMSE (dB)")
plt.gca().invert_yaxis()  # better NMSE (more negative) at top
plt.xticks(epochs)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
